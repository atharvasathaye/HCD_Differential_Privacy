```python
# 311 Service Requests — Local CSV EDA (Notebook-only, no file outputs)
# Dependencies: pandas, matplotlib (and numpy)
# How to use: set PATH to your CSV and run this single cell.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# ============== Config ==============
PATH = "/311_Service_Requests_small.csv"  # change if your CSV is elsewhere

# friendlier DataFrame display
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 120)

# Matplotlib basics (no specific styles/colors to keep it portable)
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.dpi"] = 120


# ============== Helpers ==============
def safe_to_datetime(df: pd.DataFrame, candidates):
    """
    Try converting the given candidate columns to datetime; return the first that works.
    If none, heuristically try any column containing 'date' in its name.
    """
    for c in candidates:
        if c in df.columns:
            try:
                df[c] = pd.to_datetime(df[c], errors="coerce", utc=False)
                if df[c].notna().any():
                    return c
            except Exception:
                pass
    for c in df.columns:
        if "date" in c.lower():
            try:
                df[c] = pd.to_datetime(df[c], errors="coerce", utc=False)
                if df[c].notna().any():
                    return c
            except Exception:
                pass
    return None


def first_present(df: pd.DataFrame, candidates):
    """Return the first candidate column that exists in df, else None."""
    for c in candidates:
        if c in df.columns:
            return c
    return None


def describe_all_safe(df: pd.DataFrame) -> pd.DataFrame:
    """
    Backward-compatible 'describe' that:
      - uses df.describe() for numeric
      - uses df.describe() for object/category/bool
      - builds a custom summary for datetime columns (count/min/max/nunique)
    """
    # numeric
    num = df.select_dtypes(include=[np.number])
    num_desc = num.describe().T if not num.empty else pd.DataFrame()

    # object/category/bool
    obj = df.select_dtypes(include=["object", "category", "bool"])
    obj_desc = obj.describe().T if not obj.empty else pd.DataFrame()

    # datetime (robust detection)
    from pandas.api.types import is_datetime64_any_dtype
    dt_cols = [c for c in df.columns if is_datetime64_any_dtype(df[c])]
    dt_rows = []
    for col in dt_cols:
        s = df[col].dropna()
        if s.empty:
            row = pd.Series({"count": 0, "min": pd.NaT, "max": pd.NaT, "nunique": 0}, name=col)
        else:
            row = pd.Series(
                {"count": int(s.shape[0]), "min": s.min(), "max": s.max(), "nunique": int(s.nunique())},
                name=col,
            )
        dt_rows.append(row)
    dt_desc = pd.DataFrame(dt_rows)

    parts = []
    if not obj_desc.empty: parts.append(obj_desc)
    if not num_desc.empty: parts.append(num_desc)
    if not dt_desc.empty:  parts.append(dt_desc)
    return pd.concat(parts, axis=0, sort=False) if parts else pd.DataFrame()


# ============== 1) Read ==============
df = pd.read_csv(PATH)
print("=== Basic Info ===")
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}\n")

# ============== 2) Missing & Describe (compat) ==============
na = df.isna().sum().sort_values(ascending=False).to_frame("missing_count")
na["missing_ratio"] = (na["missing_count"] / len(df)).round(4)
display(na)  # missing value summary

desc_all = describe_all_safe(df)
display(desc_all)  # combined summary (numeric/object/bool/datetime)

# ============== 3) Detect & convert datetime columns ==============
created_col = safe_to_datetime(df, ["CREATED_DATE", "CREATE_DATE", "SR_CREATE_DATE", "CREATION_DATE"])
closed_col  = safe_to_datetime(df, ["CLOSED_DATE", "CLOSE_DATE", "SR_CLOSED_DATE", "RESOLUTION_DATE"])
print("=== Date Columns (detected) ===")
print("Created date column:", created_col)
print("Closed  date column:", closed_col, "\n")

# ============== 4) Convert common numeric fields ==============
for c in ["LATITUDE", "LONGITUDE", "WARD", "ZIP_CODE", "COMMUNITY_AREA"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ============== 5) Detect key categorical fields ==============
sr_type_col = first_present(df, ["SR_TYPE", "SERVICE_REQUEST_TYPE", "TYPE", "REQUEST_TYPE", "SR_SHORT_CODE"])
status_col  = first_present(df, ["STATUS", "SR_STATUS", "CURRENT_STATUS"])
print("=== Key Dimensions ===")
print("Service request type column:", sr_type_col)
print("Status column:", status_col, "\n")

# ============== 6) Top-N service request types ==============
if sr_type_col is not None:
    type_top = df[sr_type_col].value_counts().head(15)
    display(type_top.to_frame("count"))
    plt.figure(figsize=(10, 5))
    type_top.sort_values(ascending=True).plot(kind="barh")
    plt.title("Top 15 Service Request Types")
    plt.xlabel("Count")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()

# ============== 7) Status distribution ==============
if status_col is not None:
    status_counts = df[status_col].value_counts(dropna=False)
    display(status_counts.to_frame("count"))
    plt.figure(figsize=(8, 4))
    status_counts.plot(kind="bar")
    plt.title("Status Distribution")
    plt.xlabel("Status")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# ============== 8) Time trends + DoW/Hour (if created date exists) ==============
if created_col is not None:
    # monthly counts
    monthly = (
        df.set_index(created_col)
          .sort_index()
          .resample("M")
          .size()
    )
    print("Monthly counts (head):")
    print(monthly.head(10), "\n")
    plt.figure(figsize=(12, 4))
    monthly.plot()
    plt.title("Monthly Service Request Counts")
    plt.ylabel("Count")
    plt.xlabel("Month")
    plt.tight_layout()
    plt.show()

    # day-of-week distribution (0=Mon)
    dow = df[created_col].dt.dayofweek
    dow_counts = dow.value_counts().sort_index()
    dow_map = {0:"Mon",1:"Tue",2:"Wed",3:"Thu",4:"Fri",5:"Sat",6:"Sun"}
    dow_counts.index = [dow_map.get(i, str(i)) for i in dow_counts.index]
    display(dow_counts.to_frame("count"))
    plt.figure(figsize=(8, 4))
    dow_counts.plot(kind="bar")
    plt.title("Requests by Day of Week")
    plt.xlabel("Day")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

    # hour-of-day distribution
    hour_counts = df[created_col].dt.hour.value_counts().sort_index()
    display(hour_counts.to_frame("count"))
    plt.figure(figsize=(10, 4))
    hour_counts.plot(kind="bar")
    plt.title("Requests by Hour of Day")
    plt.xlabel("Hour (0–23)")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# ============== 9) Handling duration (if both created & closed exist) ==============
if created_col is not None and closed_col is not None:
    duration_hours = (df[closed_col] - df[created_col]).dt.total_seconds() / 3600.0
    df["_DURATION_DAYS"] = duration_hours / 24.0

    # remove negatives and extreme outliers (cap at 99th percentile)
    dur = df["_DURATION_DAYS"].replace([np.inf, -np.inf], np.nan)
    dur = dur[(dur >= 0)]
    if dur.notna().any():
        cap = dur.quantile(0.99)
        dur_cap = dur[dur <= cap]

        stats = pd.DataFrame({
            "count": [dur_cap.count()],
            "mean_days": [dur_cap.mean()],
            "median_days": [dur_cap.median()],
            "p90_days": [dur_cap.quantile(0.90)],
            "p95_days": [dur_cap.quantile(0.95)],
            "max_days(p99-capped)": [dur_cap.max()],
        })
        display(stats)

        plt.figure(figsize=(10, 4))
        plt.hist(dur_cap.dropna(), bins=40)
        plt.title("Service Request Handling Duration (days, outliers removed)")
        plt.xlabel("Duration (days)")
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()

        # means by key dimensions (if present)
        if sr_type_col is not None:
            mean_by_type = (
                df.loc[dur_cap.index]
                  .groupby(sr_type_col)["_DURATION_DAYS"]
                  .mean()
                  .sort_values(ascending=False)
            )
            display(mean_by_type.head(20).to_frame("mean_duration_days"))
            plt.figure(figsize=(10, 6))
            mean_by_type.head(15).sort_values(ascending=True).plot(kind="barh")
            plt.title("Avg Handling Duration by Request Type (days, Top 15)")
            plt.xlabel("Avg days")
            plt.ylabel("")
            plt.tight_layout()
            plt.show()

        if status_col is not None:
            mean_by_status = (
                df.loc[dur_cap.index]
                  .groupby(status_col)["_DURATION_DAYS"]
                  .mean()
                  .sort_values(ascending=False)
            )
            display(mean_by_status.to_frame("mean_duration_days"))
            plt.figure(figsize=(8, 4))
            mean_by_status.sort_values(ascending=True).plot(kind="barh")
            plt.title("Avg Handling Duration by Status (days)")
            plt.xlabel("Avg days")
            plt.ylabel("")
            plt.tight_layout()
            plt.show()

        for geo_col, label in [("WARD", "WARD"), ("COMMUNITY_AREA", "COMMUNITY_AREA"), ("ZIP_CODE", "ZIP_CODE")]:
            if geo_col in df.columns:
                mean_by_geo = (
                    df.loc[dur_cap.index]
                      .groupby(geo_col)["_DURATION_DAYS"]
                      .mean()
                      .sort_values(ascending=False)
                )
                display(mean_by_geo.head(20).to_frame("mean_duration_days"))
                plt.figure(figsize=(10, 6))
                mean_by_geo.head(20).sort_values(ascending=True).plot(kind="barh")
                plt.title(f"Avg Handling Duration by {label} (days, Top 20)")
                plt.xlabel("Avg days")
                plt.ylabel("")
                plt.tight_layout()
                plt.show()

print("✅ Done — results are displayed inline above.")
```
